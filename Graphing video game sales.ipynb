```python
import seaborn as sns
import pandas as pd
sns.set_theme()
import matplotlib.pyplot as plt


game_sales = pd.read_csv("top100videogamesales.csv")
game_sales.head()
```

# 1. Regional Sales by Publisher:

### North American Sales by publisher

```python
NA_sales = game_sales.groupby(["Publisher"]).sum().plot(kind="bar", title='Regional sales by publisher(NA_Sales)', color = "Grey", y="NA_Sales")
```

#### NOTE: Nintendo had the most NA Sales, while saga had the least

```python
EU_sales = game_sales.groupby(["Publisher"]).sum().plot(kind="bar", title='Regional sales by publisher(EU_Sales)', color = "Orange", y="EU_Sales")
```

# 4. North American Sales by Genre:

```python
import matplotlib.pyplot as plt

Action = game_sales[game_sales["Genre"] == "Action"].NA_Sales.sum().round(2)

Sports = game_sales[game_sales["Genre"] == "Sports"].NA_Sales.sum().round(2)

Racing = game_sales[game_sales["Genre"] == "Racing"].NA_Sales.sum().round(2)

Other = game_sales[(game_sales["Genre"] == "Fighting") | (game_sales["Genre"] == "Puzzle") | (game_sales["Genre"] == "Simulation")].NA_Sales.sum().round(2)

Misc = game_sales[game_sales["Genre"] == "Misc"].NA_Sales.sum().round(2)

Platform = game_sales[game_sales["Genre"] == "Platform"].NA_Sales.sum().round(2)

Role_Playing = game_sales[game_sales["Genre"] == "Role-Playing"].NA_Sales.sum().round(2)

Shooter = game_sales[game_sales["Genre"] == "Shooter"].NA_Sales.sum().round(2)

Data = [Action, Misc, Platform, Racing, Role_Playing, Shooter, Sports, Other]
Label = ["Action", "Misc", "Platform", "Racing", "Role_Playing", "Shooter", "Sports", "Other"]
Color = ["#C11B17", "#E77471", "#E56717", "#FD1C03", "#E75480", "#A74AC7", "#0000A5", "#B0BF1A"] 


plt.pie(Data, labels = Label, colors = Color, autopct = "%1.1f%%", startangle = 90)
plt.title("NA_Sales by genre", fontsize = 16)
plt.axis("equal")
plt.show()
```

#### Shooter games had the most percentage in the North American Market.

# 5. European Sales by Genre:

```python
import matplotlib.pyplot as plt

Action = game_sales[game_sales["Genre"] == "Action"].EU_Sales.sum().round(2)

Sports = game_sales[game_sales["Genre"] == "Sports"].EU_Sales.sum().round(2)

Racing = game_sales[game_sales["Genre"] == "Racing"].EU_Sales.sum().round(2)

Other = game_sales[(game_sales["Genre"] == "Fighting") | (game_sales["Genre"] == "Puzzle") | (game_sales["Genre"] == "Simulation")].EU_Sales.sum().round(2)

Misc = game_sales[game_sales["Genre"] == "Misc"].EU_Sales.sum().round(2)

Platform = game_sales[game_sales["Genre"] == "Platform"].EU_Sales.sum().round(2)

Role_Playing = game_sales[game_sales["Genre"] == "Role-Playing"].EU_Sales.sum().round(2)

Shooter = game_sales[game_sales["Genre"] == "Shooter"].EU_Sales.sum().round(2)

Data = [Action, Misc, Platform, Racing, Role_Playing, Shooter, Sports, Other]
Label = ["Action", "Misc", "Platform", "Racing", "Role_Playing", "Shooter", "Sports", "Other"]
Color = ["#B666D2", "#893BFF", "#5453A6", "#F433FF", "#E4287C", "#FFCCCB", "#F8B88B", "#7F5A58"]


plt.pie(Data, labels = Label, colors = Color, autopct = "%1.1f%%", startangle = 90)
plt.title("EU_Sales by genre", fontsize = 16)
plt.axis("equal")
plt.show()
```

# (ii) What percentage of global sales does PlayStation have in the dataset?

```python
import matplotlib.pyplot as plt

Nintendo = game_sales[game_sales["Platform"] == "Nintendo"].Global_Sales.sum().round(2)

XBox = game_sales[game_sales["Platform"] == "XBox"].Global_Sales.sum().round(2)

PlayStation = game_sales[game_sales["Platform"] == "PlayStation"].Global_Sales.sum().round(2)

Other = game_sales[game_sales["Platform"] == "Other"].Global_Sales.sum().round(2)

Data = [Nintendo, XBox, PlayStation, Other]
Label = ["Nintendo", "XBox", "PlayStation", "Other"]
Color = ["#915F6D", "#583759", "Grey", "#0000CD"]

plt.pie(Data, labels = Label, colors = Color, autopct = "%1.1f%%")
plt.title("Percentage of platforms in global sales", fontsize = 16)
plt.axis("equal")
plt.show()
```

#### PlayStation contains 27.1% in the global sales
